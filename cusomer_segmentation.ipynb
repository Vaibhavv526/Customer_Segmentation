```python
cluster_names = {
    0: "Budget Customers",
    1: "Premium Customers",
    2: "Loyal High Spenders",
    3: "Regular Customers",
    4: "Occasional Buyers",
    5: "High Income Low Spenders"
}

df["Cluster_Name"] = df["Clusters"].map(cluster_names)
```
